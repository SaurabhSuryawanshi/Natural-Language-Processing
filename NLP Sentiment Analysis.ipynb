```python
ds['Text']=ds['Text'].str.lower()
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
    if isinstance (text,str):
        tokens=word_tokenize(text)
        filtered_tokens=[lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and word.isalnum()]
        return ' '.join(filtered_tokens)
        return np.nan

ds['Text']=ds['Text'].apply(preprocess_text)
```

```python
encoder=OneHotEncoder()
sentiment_encoded=encoder.fit_transform(ds[['Sentiment']])
```

```python
source_encoded = encoder.fit_transform(ds[['Source']]).toarray()
userid_encoded = encoder.fit_transform(ds[['User ID']]).toarray()
location_encoded=encoder.fit_transform(ds[['Location']]).toarray()
```

```python
print(encoder.feature_names_in_)
```

```python
encoder.fit(ds[['Source']])
source_encoded = encoder.transform(ds[['Source']]).toarray()
source_ds = pd.DataFrame(source_encoded, columns=encoder.get_feature_names_out(['Source']))
```

```python
encoder.fit(ds[['User ID']])
userid_encoded = encoder.transform(ds[['User ID']]).toarray()
userid_ds = pd.DataFrame(userid_encoded, columns=encoder.get_feature_names_out(['User ID']))
```

```python
ds.tail()
```

```python
encoder.fit(ds[['Location']])
location_encoded = encoder.transform(ds[['Location']]).toarray()
location_ds = pd.DataFrame(location_encoded, columns=encoder.get_feature_names_out(['Location']))
```

```python
ds = pd.concat([ds, source_ds, userid_ds,  location_ds], axis=1)
print(ds.shape)
```

```python
ds['Date/Time'] = pd.to_datetime(ds['Date/Time'], errors='coerce')
ds['Year'] = ds['Date/Time'].dt.year
```

```python
scaler = MinMaxScaler()
numerical_columns = ['Confidence Score']
ds[numerical_columns] = scaler.fit_transform(ds[numerical_columns])
```
